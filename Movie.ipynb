```python
# 1. Create subsets according to original language

data['original_language'].value_counts()
```

```python
languages = list(data['original_language'].unique())
```

```python
subsets = []

for n in range(len(languages)):
    languages[n] = pd.DataFrame(data[data['original_language'] == languages[n]])
    subsets.append(languages[n])
```
